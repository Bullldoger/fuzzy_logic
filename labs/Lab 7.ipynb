```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt
```

```python
alfa_x = 0.2
betta_x = 0.3
a_x = 0.5
```

```python
alfa_y = 0.2
betta_y = 0.3
a_y = 0.3
```

# Основной блок кода с арифметическими функциями

```python
def arithmetic(alfa_x, betta_x, a_x, alfa_y, betta_y, a_y, operation : str):
    
    data_x = np.linspace(a_x - alfa_x, a_x + betta_x, 1000)
    x = fuzz.trimf(data_x, [a_x - alfa_x, a_x, a_x + betta_x])
    
    data_y = np.linspace(a_y - alfa_y, a_y + betta_y, 1000)
    y = fuzz.trimf(data_y, [a_y - alfa_y, a_y, a_y + betta_y])
    
    def sum():
        
        a_z = a_x + a_y
        alfa_z = alfa_x + alfa_y
        betta_z = betta_x + betta_y
        
        data_z = np.linspace(a_z - alfa_z, a_z + betta_z, 1000)
        z = fuzz.trimf(data_z, [a_z - alfa_z, a_z, a_z + betta_z])
        
        return data_z, z
    
    
    def sub():
        a_z = a_x - a_y
        alfa_z = alfa_x + betta_y
        betta_z = betta_x + alfa_y
        
        data_z = np.linspace(a_z - alfa_z, a_z + betta_z, 1000)
        z = fuzz.trimf(data_z, [a_z - alfa_z, a_z, a_z + betta_z])
        
        return data_z, z
    
    
    def mul():
        
        if a_x > 0 and a_y > 0:
            a_z = a_x * a_y
            alfa_z = a_x * alfa_y + a_y * alfa_x
            betta_z = a_x * betta_y + a_y * betta_x
            
            data_z = np.linspace(a_z - alfa_z, a_z + betta_z, 1000)
            z = fuzz.trimf(data_z, [a_z - alfa_z, a_z, a_z + betta_z])
            
        elif a_x < 0 and a_y > 0:
            
            a_z = a_x * a_y
            alfa_z = a_y * alfa_x - a_x * betta_y
            betta_z = a_y * betta_x - a_x * alfa_y
            
            data_z = np.linspace(a_z - alfa_z, a_z + betta_z, 1000)
            z = fuzz.trimf(data_z, [a_z - alfa_z, a_z, a_z + betta_z])
            
        elif a_x < 0 and a_y < 0:
            
            a_z = a_x * a_y
            alfa_z = - a_y * betta_x - a_x * betta_y
            betta_z = - a_y * alfa_x - a_x * alfa_y
#             !!!!!
            
            data_z = np.linspace(a_z - alfa_z, a_z + betta_z, 1000)
            z = fuzz.trimf(data_z, [a_z - alfa_z, a_z, a_z + betta_z])
        
        return data_z, z
    
    
    def div():
        
        if a_x > 0 and (a_x - alfa_x) > 0 and a_y > 0 and (a_y - alfa_y) > 0:
            a_z = a_x / a_y
            alfa_z = (a_x * betta_y + a_y * alfa_x) / pow(a_y, 2)
            betta_z = (a_x * alfa_y + a_y * betta_x) / pow(a_y, 2)
            
            data_z = np.linspace(a_z - alfa_z, a_z + betta_z, 1000)
            z = fuzz.trimf(data_z, [a_z - alfa_z, a_z, a_z + betta_z])
            
            return data_z, z
        
        else:
            print('Exeption!')
            
            
    if operation == 'sum':
        data_z, z = sum()
        
    elif operation == 'sub':
        data_z, z = sub()
        
    elif operation == 'mul':
        data_z, z = mul()
        
    elif operation == 'div':
        data_z, z = div()
        
    return data_x, x, data_y, y, data_z, z
```

## Сумма

```python
answer1 = arithmetic(alfa_x, betta_x, a_x, alfa_y, betta_y, a_y, operation='sum')
```

```python
plt.figure(figsize=(10,7))
plt.plot(answer1[0], answer1[1], label='x')
plt.plot(answer1[2], answer1[3], label='y')
plt.plot(answer1[4], answer1[5], label='z')
plt.legend()
plt.grid()
plt.xlabel('значение')
plt.ylabel('степень принадлежности')
plt.show()
```

## Вычитание
