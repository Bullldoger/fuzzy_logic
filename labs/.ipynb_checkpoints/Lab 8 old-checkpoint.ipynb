## Fuzzy filters (ATMAV, ATMED)
* ATMED - Asymmetrical Triangular Median Filter.
* ATMAV - Asymmetrical Triangular Moving Average Filter

```python
def ATMED(value, v_min, v_median, v_max):
    
    if v_min <= value <= v_median:
        return 1 - (v_median - value) / (v_median - v_min)
    
    if v_median <= value <= v_max:
        return 1 - (value - v_median) / (v_max - v_median)
    
    if v_min == v_median or v_median == v_max:
        return 1
```

```python
def ATMAV(value, v_min, v_mean, v_max):
    
    if v_min <= value <= v_mean:
        return 1 - (v_mean - value) / (v_mean - v_min)
    
    if v_mean <= value <= v_max:
        return 1 - (value - v_mean) / (v_max - v_mean)
    
    if v_min == v_mean or v_mean == v_max:
        return 1
```

```python
def construct_filter(sub_matrix, filter_name='ATMAV'):
    
    row = sub_matrix.ravel().astype(np.float64)
    
    v_min = np.min(row)
    v_max = np.max(row)
    
    v_median = np.median(row)
    v_mean = np.mean(row)

    func = None
    
    if filter_name == 'ATMAV':
        v_m = v_mean
        func = ATMAV
    if filter_name == 'ATMED':
        v_m = v_median
        func = ATMED
        
    FILTER = np.zeros(sub_matrix.shape)
    
    h, w = FILTER.shape
    
    for i in range(h):
        for j in range(w):
            value = sub_matrix[i, j]
            FILTER[i, j] = func(value, v_min, v_m, v_max)
            
    return FILTER
```

```python
def fuzzy_filter(img, window_size=12, method='ATMAV'):

    h, w = img.shape
    
    output = np.zeros(img.shape)
    
    output[0, :] = img[0, :]
    output[:, 0] = img[:, 0]
    output[-1, :] = img[-1, :]
    output[:, -1] = img[:, -1]
    
    for i in range(window_size, h - window_size - 1):
        for j in range(window_size, w - window_size - 1):
            sub_img = img[i - window_size: i + window_size + 1, j - window_size: j + window_size + 1]
            FILTER = construct_filter(sub_matrix=sub_img, filter_name=method)
            
            output[i, j] = (sub_img * FILTER).ravel().sum()
            
    return output
```

# Results

```python
def util(blurred):
    coeffs = pywt.dwt2(blurred, 'haar')
    cA, (cH, cV, cD) = coeffs
    cA = fuzzy_filter(cA, method='ATMAV', window_size=1)
    cH = np.clip(cH, -10, 10)
    cV = np.clip(cV, -10, 10)
    cD = np.clip(cD, -10, 10)
    r = np.zeros(cD.shape)
    atmav_recovered = pywt.idwt2((cA, (cH, cV, cD)), 'haar')
    
    coeffs = pywt.dwt2(blurred, 'haar')
    cA, (cH, cV, cD) = coeffs
    cA = fuzzy_filter(cA, method='ATMED', window_size=1)
    cH = np.clip(cH, -10, 10)
    cV = np.clip(cV, -10, 10)
    cD = np.clip(cD, -10, 10)
    r = np.zeros(cD.shape)
    atmed_recovered = pywt.idwt2((cA, (cH, cV, cD)), 'haar')
    
    wiener_recovered = scipy.signal.wiener(np.array(blurred))
    
    return atmed_recovered, atmav_recovered, wiener_recovered
```

```python
def plot_results(blurred, atmav_recovered, atmed_recovered, wiener_recovered):
    plt.figure(figsize=(10, 8))

    plt.subplot(221)
    plt.title('Изначально зашумленное изображение')
    plt.imshow(blurred, cmap='gray')

    plt.subplot(222)
    plt.title('Fuzzy ATMAV')
    plt.imshow(atmav_recovered, cmap='gray')

    plt.subplot(223)
    plt.title('Fuzzy ATMED')
    plt.imshow(atmed_recovered, cmap='gray')

    plt.subplot(224)
    plt.title('Wiener filter')
    plt.imshow(wiener_recovered, cmap='gray')

    plt.show()
```

```python
%time atmed_recovered, atmav_recovered, wiener_recovered = util(blurred_1)
plot_results(blurred_1, atmed_recovered, atmav_recovered, wiener_recovered)
```

# Image contrast enhancment by fuzzy inference system

From "Image Enhancement Using MamdaniFuzzy Inference System":[fuzzy_logic/Labs and lectures/Image Enhancement Using MamdaniFuzzy Inference System.pdf]

```python
def fuzzy_contrast_enhancer(data, FE=255, FD=2, window=2):
    rows, cols = data.shape
    P_MEAN = np.copy(data)
    P_MAX = np.copy(data)
    P_MIN = np.copy(data)
    P_MEDIAN = np.copy(data)
    P_STD = np.copy(data)
    
    dv = window // 2
    
    for i in range(dv, rows - dv - 1):
        
        for j in range(dv, cols - dv - 1):
            moveable_window = data[i - dv: i + dv + 1, j - dv: j + dv + 1]

            P_MEAN[i, j] = np.mean(moveable_window)
            P_MAX[i, j] = np.max(moveable_window)
            P_MIN[i, j] = np.min(moveable_window)
            P_MEDIAN[i, j] = np.median(moveable_window)
            P_STD[i, j] = np.var(moveable_window)
            
    return P_MEAN, P_MAX, P_MIN, P_MEDIAN, P_STD
```

## Mandami/Sugeno FIS

```python
pixel = ctrl.Antecedent(np.arange(0, 1, 0.1), 'pixel')
new_pixel_mandami = ctrl.Consequent(np.arange(-0.5, 0.5, 0.1), 'new_pixel_mandami')

pixel['bright'] = fuzz.trimf(pixel.universe, [0, 0, 0.5])
pixel['gray'] = fuzz.trimf(pixel.universe, [0, 0.5, 1])
pixel['dark'] = fuzz.trimf(pixel.universe, [0.5, 1, 1])

new_pixel_mandami['brighter'] = fuzz.trimf(new_pixel_mandami.universe, [-0.5, -0.25, 0])
new_pixel_mandami['gray'] = fuzz.trimf(new_pixel_mandami.universe, [-0.25, 0, 0.25])
new_pixel_mandami['darker'] = fuzz.trimf(new_pixel_mandami.universe, [0, 0.25, 0.5])

rule_mandami_1 = ctrl.Rule(pixel['bright'], new_pixel_mandami['brighter'])
rule_mandami_2 = ctrl.Rule(pixel['gray'], new_pixel_mandami['gray'])
rule_mandami_3 = ctrl.Rule(pixel['dark'], new_pixel_mandami['darker'])

pixels_mandami_ctrl = ctrl.ControlSystem([rule_mandami_1, 
                                     rule_mandami_2, 
                                     rule_mandami_3])

pixels_mandami = ctrl.ControlSystemSimulation(pixels_mandami_ctrl)
```

```python
pixels_mandami.input['pixel'] = 0.4
pixels_mandami.compute()

print(pixels_mandami.output['new_pixel_mandami'])
# pixels_mandami.view(sim=pixels_mandami)
```

```python
def convert_data_to_fis(P):
    rows, cols = P.shape
    
    P_M = np.copy(P)
    
    for i in range(rows):
        for j in range(cols):
            value = P[i, j] / 255
            v = value
            pixels_mandami.input['pixel'] = v
            pixels_mandami.compute()
            P_M[i, j] = (P[i, j] + pixels_mandami.output['new_pixel_mandami']) * 255           
    return P_M
```

```python
%time P_MEAN, P_MAX, P_MIN, P_MEDIAN, P_STD_1 = fuzzy_contrast_enhancer(pict[:, :, 0])
%time P_MEAN, P_MAX, P_MIN, P_MEDIAN, P_STD_2 = fuzzy_contrast_enhancer(pict[:, :, 1])
%time P_MEAN, P_MAX, P_MIN, P_MEDIAN, P_STD_3 = fuzzy_contrast_enhancer(pict[:, :, 2])
%time P_MEAN, P_MAX, P_MIN, P_MEDIAN, P_STD_4 = fuzzy_contrast_enhancer(pict[:, :, 3])
```
